# Dataset criado pelo professor:

```python
import numpy as np
import torch
from torch.utils.data import TensorDataset
from sklearn.datasets import load_iris

iris = load_iris()
data = iris.data[iris.target==1, ::2]  # sépala (coluna 0) e pétala (coluna 2)

x_data = data[:,0:1]
target = data[:,1:2]

# Convertendo para tensores
x_data = torch.FloatTensor(x_data)
target = torch.FloatTensor(target)

# Criando o Dataset usando TensorDataset
dataset = TensorDataset(x_data, target)

print("Tamanho do dataset:", len(dataset))
print("Exemplo:", dataset[0])
```

# Atividade:

##Defina seu próprio DataLoader na segunda célula abaixo. Faça com que cada batch tenha 5 elementos cada. Utilize o dataset já criado anteriormente. Em seguida imprima o tamanho do Dataloader e tamanho do Dataset

```python
from torch.utils.data import DataLoader

data_loader = DataLoader(dataset,
                         batch_size=5,
                         shuffle=False)

print("tamanho do DataLoader:", len(data_loader))
print("tamanho do dataset:", len(dataset))
```

## A ordem dos elementos se altera entre execuções da célula acima?


```python
print("Batchs nesta execução:")

for batch in data_loader:
    print(batch)
```

Com shuffle=False a ordem não muda entre execuções da célula

## Tente colocar o parametro shuffle do DataLoader como True e obeseve o resultado do exercício, rodando várias vezes.

```python
data_loader_shuffle = DataLoader(dataset,
                                 batch_size=5,
                                 shuffle=True)

print("Batchs com shuffle=True:")

for batch in data_loader_shuffle:
    print(batch)
```
